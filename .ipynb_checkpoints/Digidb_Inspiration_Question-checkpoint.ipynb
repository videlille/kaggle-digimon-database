# Digimon Database - Vi 

Context:
Digimon, short for “digital monsters”, is a franchise which revolves around a core mechanic of capturing, caring for and training monsters and then engaging in combat with them. It’s similar to Pokémon.

This dataset contains information on digimon from “Digimon Digimon Story: Cyber Sleuth”, released for Playstation Vita in 2015 and Playstation 4 in 2016.

Content: 

This database contains three files: 
- DigiDB_digimonlist: a list of 249 the digimon along with their stage, typo, attribute, memory, and number of equipment slots. All stats (HP, SP, attack, defense, intelligence and speed) are for that Digimon at level 50.
- DigiDB_movelist: all the moves which Digimon can perform
- DigiDB_supportlist: all the Support Skills. (Support Skills are a passive, stackable, team-wide buff. Each species of Digimon is associated with a single Support Skill.)

Questions:

1) Which set of moves will get the best ratio of attack power to SP spent?
2) Which team of 3 digimon have the highest attack? Defense?
3) What’s the tradeoff between HP and SP?
4) Are some types over- or under-represented?
5) Both the moves and support skills have short text descriptions. Can an NLP analysis reveal underlying clusters of moves?
6) Are different types and attributes evenly represented across stages?

Source: https://www.kaggle.com/rtatman/digidb

Note: The words Attributes and Types are used in different fandom websites to designated different Digimon characteristics. In the scope of this wok, we will stick to the exact vocabulary used in the questions and in the database.

## Question 3: What’s the tradeoff between HP and SP?


```python
import seaborn as sns
import numpy as np

sns.scatterplot(x="Lv 50 HP", y="Lv50 SP", data=digimonList)
```

As the direct relation between Lv50 SP and Lv50 HP is not obvious, let's plot that graph against an other stat variable

```python
fig, axes = plt.subplots(nrows=2, ncols=2,constrained_layout=True)

digimonList.plot.scatter(x='Lv 50 HP',y='Lv50 SP',c='Lv50 Atk',colormap='viridis',ax=axes[0,0])
digimonList.plot.scatter(x='Lv 50 HP',y='Lv50 SP',c='Lv50 Def',colormap='viridis',ax=axes[0,1])
digimonList.plot.scatter(x='Lv50 Int',y='Lv50 SP',c='Lv 50 HP',colormap='viridis',ax=axes[1,0])
digimonList.plot.scatter(x='Lv 50 HP',y='Lv50 SP',c='Lv50 Spd',colormap='viridis',ax=axes[1,1])
```

```python
sns.lmplot( x='Lv 50 HP', y='Lv50 SP', data=digimonList, fit_reg=False, hue='Stage', legend=False)
plt.legend(loc='lower right')
plt.show()
```

A general trend appear on that graph. Hp-SP is dependent on the Digimon Stage. Generally, the low SP-HP couple Digimon are mostly Baby and In-Training. Middle-range HP-SP are associated to Rookie and champion. The upper values are for Ultimate, Mega, Ultra and Armor. However, those categories don't have clear borders and they overlap on each other. A Rookie ( a green dot around 150 SP ) can have better HP-SP stats than some Mega digimons. 


Further analysis could be done by a linear interpolation of HP-SP for each category and compared by the slope to observe the magnitude of the trade-off. A visual analysis tell us that the slope is negative for each category, it means that as HP are higher, the SP is lower. 

Similar plots can be displayed with the different types and Attributes as color range

```python
sns.lmplot( x='Lv 50 HP', y='Lv50 SP', data=digimonList, fit_reg=False, hue='Type', legend=False)
plt.legend(loc='lower right')
plt.show()

sns.lmplot( x='Lv 50 HP', y='Lv50 SP', data=digimonList, fit_reg=False, hue='Attribute', legend=False)
plt.legend(loc='lower right')
plt.show()
```

## Question 5: Both the moves and support skills have short text descriptions. Can an NLP analysis reveal underlying clusters of moves?

Let's have a close look to the move description. 


A lot of attack descriptions in MoveList follow the pattern ' [Type] attack, 65 [Attribute] damage to [one or all] foe. ' Those two variables are already described in the database and don't bring additional informations. Some have additional notes such as '95% accuracy.'/'Paralyze status effect.'/..  which are not captured by the other columns.

Others descriptions in MoveList and supportList have different action verbs as Recover\Stunt\Remove\Increases... , act on different parameters as INT\DEF\ATK\... with a certain magnitude and can apply to different target. Using NLP to perform Keywords extractions could help clusterize those moves by new different categories as Healing, Protection, Speed Boost or even categories the analyst wouldn't have tought of.


```python
moveList.head()
```

For the sake of simplicity, An external NLP app, MonkeyLearn is used for keywords extraction. This is the features extraction. The keywords from the description are listed in a new column.

```python
moveList_Extract_Keywords = pd.read_csv('DigiDB_movelist_Keywords.csv', sep=';')
#print(moveList_Extract_Keywords.shape)
moveList_Extract_Keywords.sample(frac=1).head(5)
```

By going through the keywords, we ca see that Keyword extraction shows underlying concepts that don't appear in the other columns as 'foe', 'paralyze status', 'stun status' or 'enemy'.  Using those keywords as tag permits new classifications of moves. To improve the results, a home-made keyword extraction model can be implemented and trained using directly our descriptions to identify better keywords.

At this stage, two scenarios are possible:
- Moves classification based on keywords that act as tags or category
- Clusterization, an unsupervised learning algorithm, that bundles those keywords into a pre-defined number of clusters. 


### Question 5 - Appendix - Clusterization with Kmeans

If we decide to spend a bit more time on that question, let's see if we can use a clusterization algorithm on our descriptions to identify some clusters. The steps here are:
- Divide the moves into train and test set
- Vectorize our keywords
- Use a k-mean algorithm as model and train it
- Identify clusters and interpretate them
- Prediction

```python
random_state=42
df=moveList_Extract_Keywords[['Move','Description','KEYWORD']].copy()
train=df.sample(frac=0.8,random_state=random_state)
test=df.drop(train.index)

test.reset_index(inplace=True) 

print(' Train shape is ' + str(train.shape))
print(' Test shape is ' + str(test.shape))
```
